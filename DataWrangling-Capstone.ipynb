```python
#Import pandas,json and numpy
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="whitegrid", color_codes=True)
```

```python
# read the credit_card_default file and get the data in a pandas dataframe
df = pd.read_csv("credit_card_default.csv")
print(df.head(10))
df.info()
df.describe()
```

```python
# Perform a boxplot for the columns which depicts number of past payments due
cols = ['PAY_0','PAY_2','PAY_3','PAY_4','PAY_5','PAY_6']
df[cols].plot(kind='box', subplots= True)
plt.show()
```

```python
# Perform a violin plot of number of past payments due using seaborn

sns.violinplot(x='SEX', y='PAY_0', data = df)
plt.show()

sns.violinplot(x='SEX', y='PAY_2', data = df)
plt.show()

sns.violinplot(x='SEX', y='PAY_3', data = df)
plt.show()

sns.violinplot(x='SEX', y='PAY_4', data = df)
plt.show()

sns.violinplot(x='SEX', y='PAY_5', data = df)
plt.show()

sns.violinplot(x='SEX', y='PAY_6', data = df)
plt.show()

```

```python
# convert categirocal variables to string
df['SEX'] = df['SEX'].map({1: 'male', 2: 'female'})
df.head()
```

```python
df['EDUCATION'] = df['EDUCATION'].map({1: 'graduate school', 2: 'university', 3:'high school', 4:'others'})
df.head(10)
```

```python
# Create a new Age range column
df['AGE_Range'] = pd.cut(df.AGE, bins=[20,29,39,49,59,69,79], labels=['20-29','30-39','40-49','50-59','60-69','70-79'])
df.head(10)
```

```python
df['MARRIAGE'] = df['MARRIAGE'].map({1: 'married', 2: 'single', 3:'others'})
df.head(10)
```

```python
#
df['AVG_6MNTH_BAL'] = (df['BILL_AMT1'] + df['BILL_AMT2'] + df['BILL_AMT3'] + df['BILL_AMT4'] + df['BILL_AMT5'] + df['BILL_AMT6'])/6
df.head()
```

```python
# Max missed payments
#cols = ['PAY_0','PAY_2','PAY_3','PAY_4','PAY_5','PAY_6']
df['MISSED_PYMTS'] = df[['PAY_0','PAY_2','PAY_3','PAY_4','PAY_5','PAY_6']].max(axis=1) 
df['BALANCE_TO_LIMIT'] = df['AVG_6MNTH_BAL']/df['LIMIT_BAL']
df.head()
```

```python
# Create a dataframe with columns that are needed for analysis
new_df = df.filter(['ID','LIMIT_BAL','SEX','EDUCATION','MARRIAGE','AGE_Range','MISSED_PYMTS','BALANCE_TO_LIMIT','default payment next month'], axis=1)
new_df.head(10)
```

```python
# Out of 30k, how many defaulted next month
pd.value_counts(df['default payment next month'],sort=False)
```

```python
''' High School Graduates tend to default the most; whereas graduate school people tend to default the least.
So, it is clear that chance of default decreases as education increases
'''
new_df.groupby('EDUCATION')['default payment next month'].mean()
```

```python
''' High School Graduates tend to default the most; whereas graduate school people tend to default the least.
So, it is clear that chance of default decreases as education increases
Even in all education levels males tend to default more than females.
'''
new_df.groupby(['EDUCATION','SEX'])['default payment next month'].mean()
```

```python
# Credit limit increases as age increases
new_df.groupby('AGE_Range')['LIMIT_BAL'].median()
```

```python
'''
If we ignore the 'Others' part in education, females with University education in the age range of 70 to 79 
tend to have the highest credit limit. In general, lower the education lower the credit limit.
But, males with high school or university education in 20 to 29 age range have lowest credit limit. 
Surprisingly, females with graduate school education and 70 to 79 age range also have low credit limit.
'''
new_df.groupby(['AGE_Range','SEX','EDUCATION'])['LIMIT_BAL'].median()
```

```python
new_df.groupby(['AGE_Range','SEX','EDUCATION'])['BALANCE_TO_LIMIT'].mean()
```
